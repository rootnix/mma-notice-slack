```python
import requests
from bs4 import BeautifulSoup
import shlex
```

```python
r = requests.post(
    'https://work.mma.go.kr/caisBYIS/board/boardList.do',
    data={
        'gesipan_gbcd': 13,
        'tmpl_id': 1,
        'menu_id': 'm_m8_6',
        'pageUnit': 999999,
        'pageIndex': 1
    }
)

print(r.text)
```

```python
soup = BeautifulSoup(r.text, 'lxml')
```

```python
total_page = int(soup.select_one('div.topics').text.split('/')[1].split(' ')[0])
total_page
```

```python
serial_numbers = [shlex.split(link['onclick'].split('(')[1].split(')')[0].replace(',', ' '))[2] for link in soup.select('td a')]
```

```python
for serial_number in serial_numbers[:10]:
    r = requests.post(
        'https://work.mma.go.kr/caisBYIS/board/boardView.do',
        data={
            'menu_id': 'm_m8_6',
            'gesipan_gbcd': 13,
            'ilryeon_no': serial_number,
            'searchCondition': '', 
            'searchKeyword': '',
            'pageIndex': 2,
            'pageUnit': 10,
        }
    )

    soup = BeautifulSoup(r.text)
    
    title, writer, write_date, view, attachments, content = soup.select('#boardViewform td')
    print(title.text.strip(), writer.text.strip(), write_date.text.strip(), view.text.strip())
```
